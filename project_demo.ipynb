```python
# explanation_chain.py
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Initialize OpenAI API (you can use HuggingFace too)
llm = OpenAI(temperature=0.7)

def get_explanation(topic, user_level):
    prompt = PromptTemplate(
        input_variables=["topic", "level"],
        template="""
        You are a helpful learning assistant.
        Explain the topic "{topic}" to a {level}-level student in a clear and easy way.
        Make it simple and engaging.
        """
    )
    chain = LLMChain(llm=llm, prompt=prompt)
    return chain.run({"topic": topic, "level": user_level})
```

```python
import ipywidgets as widgets
from IPython.display import display, Markdown

# Input box for user query
query_input = widgets.Text(
    value='',
    placeholder='Enter a topic...',
    description='Topic:',
    disabled=False
)

# Output area
output_area = widgets.Output()

# Button to trigger explanation
explain_button = widgets.Button(description="Explain")

# Function to run when button is clicked
def on_explain_clicked(b):
    output_area.clear_output()
    with output_area:
        query = query_input.value
        explanation = get_explanation(query)
        display(Markdown(f"### Explanation:\n\n{explanation}"))

explain_button.on_click(on_explain_clicked)

# Show everything
display(query_input, explain_button, output_area)
```
